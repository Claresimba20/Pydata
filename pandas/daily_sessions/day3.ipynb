```python
#Joining dataframe
df1 = pd.DataFrame({
    "ID" : [1,2,3,4],
    "Name" : ["Alice","Bob","Charlie","David"],
    "Subject" : ["Math","English","Science",'Arts']
})

df2 = pd.DataFrame({
    "ID" : [3,4,5,6],
    "Score" : [90,85,80,95],
    "Grade" : ["A","B","C","A"]
})
```

```python
df1
```

```python
df2
```

```python
#inner join
pd.merge(df1,df2,on="ID",how="inner")
```

```python
#outer join
pd.merge(df1,df2,on="ID",how="outer")
```
