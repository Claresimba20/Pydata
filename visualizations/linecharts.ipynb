```python
data['Date']= pd.to_datetime(data['Date'])
```

```python
data.info()
```

```python
data.index.name='Date'
```

```python
data['NATURAL GAS'].plot(kind='line',figsize=(12,6))
plt.title('Natural Gas Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
#plotting multiple commodities
#select the  commodities for comparison
columns_to_plot=["NATURAL GAS","WTI CRUDE","GOLD"]
#plotting the selected columns
data[columns_to_plot].plot(kind="line",figsize=(12,6))
plt.title('Comparison of selected Commodity Prices')
plt.xlabel("Date")
plt.ylabel("Price/Value")
plt.legend(title="Commodity")
plt.grid(True)
plt.show()
```

```python
#piecharts
#importing the necessary libraries
#function used=plt.pie(),it takes the parameter x whikch is the size of the pie pieces
plt.pie([1,2,3,4,5])
plt.show()
```
